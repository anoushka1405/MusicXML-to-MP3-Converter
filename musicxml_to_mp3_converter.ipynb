```python
import subprocess
import os
import threading
import tkinter as tk
from tkinter import filedialog, messagebox

import json
import os

CONFIG_FILE = "config.json"

def load_config():
    if os.path.exists(CONFIG_FILE):
        with open(CONFIG_FILE) as f:
            return json.load(f)
    return {}

def save_config(config):
    with open(CONFIG_FILE, "w") as f:
        json.dump(config, f)

config = load_config()
MUSESCORE_PATH = config.get("musescore_path", "/Applications/MuseScore 4.app/Contents/MacOS/mscore")

def convert_mxl_to_mp3(mxl_path, status_label):
    status_label.config(text="Converting... Please wait.")
    output_mp3 = os.path.splitext(mxl_path)[0] + ".mp3"
    try:
        subprocess.run([MUSESCORE_PATH, "-o", output_mp3, mxl_path], check=True)
        status_label.config(text=f"Conversion done! Saved as:\n{output_mp3}")
    except subprocess.CalledProcessError as e:
        status_label.config(text="Conversion failed.")
        messagebox.showerror("Conversion Error", str(e))

def start_conversion_thread(mxl_path, status_label):
    thread = threading.Thread(target=convert_mxl_to_mp3, args=(mxl_path, status_label))
    thread.start()

def browse_file():
    file_path = filedialog.askopenfilename(filetypes=[("MusicXML files", "*.mxl"), ("All files", "*.*")])
    if file_path:
        entry_path.delete(0, tk.END)
        entry_path.insert(0, file_path)
        status_label.config(text="")

def start_conversion():
    mxl_path = entry_path.get()
    if not mxl_path:
        messagebox.showwarning("Input Needed", "Please select a .mxl file first.")
        return
    start_conversion_thread(mxl_path, status_label)

root = tk.Tk()
root.title("MusicXML to MP3 Converter")
root.geometry("450x180")

tk.Label(root, text="Select MusicXML (.mxl) file:").pack(pady=5)
entry_path = tk.Entry(root, width=50)
entry_path.pack(pady=5)

btn_browse = tk.Button(root, text="Browse", command=browse_file)
btn_browse.pack(pady=5)

btn_convert = tk.Button(root, text="Convert to MP3", command=start_conversion)
btn_convert.pack(pady=10)

status_label = tk.Label(root, text="", fg="blue")
status_label.pack(pady=5)

root.mainloop()
```
